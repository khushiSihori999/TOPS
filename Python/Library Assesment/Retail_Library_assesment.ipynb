```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df=pd.read_csv("C://Users//Khushali//Downloads//Retail Data (1).csv")
df
```

```python
# View structure of dataset
df.info()
df.isnull().sum()
```

```python
# Shape of dataset
df.shape
```

```python
#duplicate record
df.duplicated().sum()
```

```python
#4️⃣ Missing / corrupted dates or numeric values
# df[['Order Date','Ship Date']].isnull().sum()
df['Order Date'].isnull().sum()

df.select_dtypes(include=np.number).isnull().sum()
```

```python
#5️⃣ Convert Order Date & Ship Date to datetime

df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')
df['Order Date']
```

```python
#6️ Check future / inconsistent shipping dates
df[df['Ship Date']<df['Order Date']]
```

```python
#7️Convert price columns to numeric
price_cols = ['Cost Price','Retail Price','Sub Total',
              'Discount $','Order Total','Shipping Cost','Total']

for col in price_cols:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)
    

df
```

```python
#Unique values

df['Customer Name'].unique()
df['Customer Type'].unique()
```

```python
# 9️⃣ Most common shipping modes
df['Ship Mode'].value_counts()

sns.countplot(x='Ship Mode',data=df)
plt.show()
```

```python
# 🔟 Cities with highest orders
df['City'].value_counts()
```

```python
# 11 1️⃣1️⃣ Range of quantities & prices
# df['Order Quantity'].min(), df['Order Quantity'].max()
df['Retail Price'].min(), df['Retail Price'].max()
```

```python
# 1️⃣2️⃣ Shipping duration column
df['Ship Duration']=(df['Ship Date']-df['Order Date']).dt.days
df['Ship Duration'].describe()
```

```python
# 1️⃣3️⃣ Zero or negative totals / quantity

df[(df['Total']<=0) | (df['Order Quantity']<=0)]
```

```python
# 1️⃣4️⃣ Discount % vs Discount $

df['Sub Total'] = (
    df['Sub Total']
    .astype(str)
    .str.replace(',', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace('$', '', regex=False)
    .astype(float)
)

df['Discount %'] = (
    df['Discount %']
    .astype(str)
    .str.replace('%', '', regex=False)
    .astype(float)
)
df['Discount $'] = pd.to_numeric(df['Discount $'], errors='coerce')


expected_discount = df['Sub Total'] * df['Discount %'] / 100
df[np.abs(expected_discount - df['Discount $']) > 0.01]
```

```python
# 1️⃣5️⃣ Total calculation mismatch
calculated_total = df['Sub Total'] - df['Discount $'] + df['Shipping Cost']
df[np.abs(calculated_total - df['Total']) > 0.01]
```

```python
#1️⃣6️⃣ Top 5 products by order quantity

df.groupby('Product Name')['Order Quantity'].sum().sort_values(ascending=False).head(5)
```

```python
#1️⃣7️⃣ Account Manager with highest revenue
df.groupby('Account Manager')['Total'].sum().sort_values(ascending=False).head(1)
```

```python
# 1️⃣8️⃣ Average shipping cost by mode
df.groupby('Ship Mode')['Shipping Cost'].mean()
```

```python
# 1️⃣ Total revenue across all orders
df['Total'].sum()
```

```python
# 2️⃣ Customer type generating more revenue
df.groupby('Customer Type')['Total'].sum().sort_values(ascending=False)
```

```python
# 3️⃣ Impact of order priority on revenue
df.groupby('Order Priority')['Total'].sum().sort_values(ascending=False)
```

```python
# 4️⃣ Average profit margin by product category

df['Profit Margin'].dtype


df['Profit Margin'] = (
    df['Profit Margin'].astype(str).str.replace('%', '', regex=False).astype(float)
)
df['Profit Margin'] = (
    df['Profit Margin']
    .astype(str)
    .str.replace('%', '', regex=False)
    .pipe(pd.to_numeric, errors='coerce')
)

df.groupby('Product Category')['Profit Margin'].mean().sort_values(ascending=False)
```

```python
#5️⃣ Most profitable product overall
df.groupby('Product Name')['Total'].sum().sort_values(ascending=False).head(1)
```

```python
# 6️⃣ Average shipping duration
df['Shipping Duration'].mean()
```

```python
#7️⃣ Do longer shipping times impact profit?
price_cols = ['Cost Price','Retail Price','Sub Total','Discount $',
              'Order Total','Shipping Cost','Total']
for col in price_cols:
    df[col] = df[col].replace('[\$,]', '', regex=True).astype(float)

df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
df['Shipping Duration'] = (df['Ship Date'] - df['Order Date']).dt.days
df[['Shipping Duration','Profit Margin']].corr()
```
